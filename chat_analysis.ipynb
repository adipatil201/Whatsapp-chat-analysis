```python
import re
import pandas as pd
import numpy as np
```

```python
k=open("WhatsApp Chat with Summer Goals.txt",'r',encoding='utf-8')
```

```python
chat = k.read() #chat is stored as a string
```

```python
#first thing we need to do is to make a dataframe for this data with msgs and date/time in different columns
#for that , we first will seperate our msgs from the date,time,name,:........this will be done using regex

pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s..\s-\s'        
# '\d{1,2}' ==> digit of size 1 or 2
#  '/' ==> / (slash)
# ','==> , (comma)
# ':'==> : (colon)
# '\s' ==> blank space
# '..' ==> two characters 
# '-' ==> - (hyphen)
```

```python
re.split(pattern,chat)            #removes the string which follows the variable pattern
```

```python
messages=re.split(pattern,chat)[1:]            #1st element of re.split(pattern,chat) was a '' ....so, we took all elements after it by [1:]
messages
```

```python
date_time=re.findall(pattern,chat)         #gives the strings which matches the variable pattern...../u202f corresponds to a space
date_time
```

```python
df=pd.DataFrame({'user_msgs':messages,'msg_dates':date_time})        #creating the dataframe of this data
```

```python
df.msg_dates
```

```python
#now, we will convert the string in msg_dates to datetime format.... %m-month,%d=day,%y=year(2 digits),%Y=year(4 digits),%I=hours(12 hr format),%H=hours(24 hr format),%M=minutes,%p=AM/PM
pd.to_datetime(df.msg_dates,format='%m/%d/%y, %I:%M %p - ')
```

```python
df.msg_dates=pd.to_datetime(df.msg_dates,format='%m/%d/%y, %I:%M %p - ')
```

```python
df
```

```python
df.shape       
#tell us the no of msgs in the entire chat
```

```python
#now, we need to seperate the user name from their msgs
users=[]
msgs=[]
for i in df.user_msgs:
    entry=re.split('([\w\W]+?):\s',i)         #if string of format pharse1-colon-pharse2 exists, then entry has ['',pharse1,pharse2]  ; else , entry = [ entire pharse]
    if entry[1:]:
        users.append(entry[1])
        msgs.append(entry[2])
    else:
        users.append('group_notif')
        msgs.append(entry[0])
    #print(entry)    
```

```python
df['users']=users
df['msgs']=msgs
```

```python
df
```

```python
df.drop(columns=['user_msgs'],inplace=True)         #droped the user_msgs column as we have seperated 
```

```python
#now , we need to first remove the group_notif as these are useless to get calculate the most common words.....
#also, we need to remove all the stopwords of both english(and,the,or,....) and hindi(kya,haa,hai,....)
#also, remove the media ommitted 
```

```python
words_df=df.msgs[df.users!='group_notif']            #removes group_notifs
words_df=words_df[words_df!='This message was deleted\n']     #removes 'This message was deleted\n'
```

```python
hinglish_stopwords_file=open('hinglish_stopwords.txt','r',encoding='utf-8')
hinglish_stopwords=hinglish_stopwords_file.read()
hinglish_stopwords=hinglish_stopwords.split()
```

```python
words=[]        
for msgs in words_df:
    k=msgs.split()
    for word in k:
        word=word.lower()   
        if word not in hinglish_stopwords:      #removes hinglish stopwords
            if word !='<Media' and word!='<media' and word != 'omitted>':   #removes media ommited
                words.append(word)
```

```python
words
```

```python
for i in Counter(words).most_common():
    print(i)
```

```python
most_common_df=pd.DataFrame(Counter(words).most_common(20))
most_common_df=most_common_df.rename(columns={0:'Words',1:'Count'})
```

```python
#now, we will plot the a heat map to see user activity for certain time in any day of the week
period=[]
for i in df.hours:
    if i==23:
        period.append(str(i)+'-'+'00')
    elif i==0:
        period.append('00'+'-'+str(i+1))
    else:
        period.append(str(i)+'-'+str(i+1))
```

```python
df['period']=period
```
